```python
import pandas as pd
```

```python
df = pd.read_csv('/content/spam (Text).csv')
```

```python
df.head()
```

```python
X = df['Message']
y = df['Category']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X_countV = CountVectorizer().fit_transform(X)
```

```python
X_countV
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_countV, y, test_size=0.2, )
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LogisticRegression()
```

```python
pred = model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y_test, pred.predict(X_test)))
```

```python
print(confusion_matrix(y_test, pred.predict(X_test)))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X_tfidf = TfidfVectorizer().fit_transform(X.values)
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2)
model = linear_model.LogisticRegression()
pred = model.fit(X_train, y_train)
print(accuracy_score(y_test, pred.predict(X_test)))
print(confusion_matrix(y_test, pred.predict(X_test)))
```
