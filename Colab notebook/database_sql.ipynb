```python
name = input("Enter name:")
```

```python
t = datetime.datetime.now()
```

```python
print(name, t)
```

```python
conn = sqlite3.connect('user.db')
c = conn.cursor()
c.execute("insert into users values(?,?)", (name, t))
conn.commit()
c.close()
conn.close()
```

```python
#select
conn = sqlite3.connect('user.db')
c = conn.cursor()
c.execute("select * from users")
r=""
for row in c:
  print(row)
  r = r+str(row)+"\n"
c.close()
conn.close()
```

```python
#delete
conn = sqlite3.connect('user.db')
c = conn.cursor()
c.execute("delete from users")
conn.commit()
c.close()
conn.close()
```
