```python
firstmsgid=0
while True:
    r=requests.get(url+f"/sendMessage?chat_id={chat_id}&text={'question: (or type quit)'}")

    while True:
        r1=requests.get(url+"/getUpdates")
        time.sleep(1)
        #print(r.json())
        r1=r1.json()
        r1=r1["result"][-1]['message']
        q=r1['text']
        msg_id=r1['message_id']

        if firstmsgid==0:
          firstmsgid=msg_id
          print(f"firstid={firstmsgid}")
          continue
        elif firstmsgid<msg_id:
          print(f"B msg={q}")
          firstmsgid=msg_id
          break

    if str.lower(q)=='quit':
        print(f"msg={q}")
        break
    else:
        r2=model.generate_content(q)
        r2=requests.get(url+f"/sendMessage?chat_id={chat_id}&text={r2.text}")
print("End")
```
